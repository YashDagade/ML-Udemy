# RANSAC Algorithm - Random Sample Consensus Algorithm



```python
X = df['RM'].values.reshape(-1,1)
y= df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X, y)
```

```python
inliner_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inliner_mask)
```

```python
np.arange(3,10,1)
```

```python
line_X = np.arange(3,10,1)
print(line_X)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
print(line_y_ransac)
```

```python
sns.set(style='whitegrid', context='notebook')
plt.figure(figsize=(12,8))
plt.scatter(X[inliner_mask], y[inliner_mask], c='blue', marker='o', label='Inliners')
plt.scatter(X[outlier_mask], y[outlier_mask], c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000s')
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

we did this for room number and medium value - let's do the same now for lstat - percent of population living in lower status and MEDV - the median value of that house

```python
X = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values
ransac_2 = RANSACRegressor()
ransac_2.fit(X, y)
inliner_mask = ransac_2.inlier_mask_
outlier_mask = np.logical_not(inliner_mask)
line_X = np.arange(0,40,1)
line_y_ransac = ransac_2.predict(line_X.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X[inliner_mask], y[inliner_mask], c='blue', marker='o', label='Inliners')
plt.scatter(X[outlier_mask], y[outlier_mask], c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('% lower status of the population')
plt.ylabel('Median value of owner-occupied homes in $1000s')
plt.legend(loc='upper left')
plt.show()
```

# Performance Evaluation of Linear Regression Models 

How do we evaluate model performance - we split the data into 2 portions. Training and Testing portion. 

```python
from sklearn.model_selection import train_test_split
```

```python
X =  df.iloc[:, :-1].values # X is now everything except the last component - so our data is everything and we are trying to predict the last column which is the MEDV
y = df['MEDV'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train) # we only train on the training data here - we are not using the test data at all
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test) # we are predicting the test data
```

# Method 1 - Residual Analysis

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10,50])
plt.show()
```

# Method 2 - Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

# Method 3 - Coefficient of Determination, $R^2$

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred) 
```

Let's see how for good Random data these values would look

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(1000)
y = 3 * x + np.random.randn(1000)
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.show()
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train.reshape(-1,1), y_train)

y_train_pred = model.predict(X_train.reshape(-1,1))
y_test_pred = model.predict(X_test.reshape(-1,1))
```

```python
# Method 1 Residual Line

plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10,50])
plt.ylim([-5,10])
plt.show()
```
